```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm
import math
import numpy as np
import statistics
from scipy.stats import skew
from scipy.stats import kurtosis
from matplotlib.ticker import FuncFormatter

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 加载数据
file_path = r'C:\Users\海棠微雨\Desktop\2022年C题\附件.xlsx'  # 数据文件路径
df = pd.read_excel(file_path, sheet_name='表单1')  # 从Excel文件中加载数据，读取名为'表单1'的工作表
```

```python
# 创建交叉表，分析风化与类型、颜色、纹饰的关系
type_wind_cross = pd.crosstab(df['类型'], df['表面风化'])  # 生成玻璃类型与表面风化的交叉表
color_wind_cross = pd.crosstab(df['颜色'], df['表面风化'])  # 生成颜色与表面风化的交叉表
pattern_wind_cross = pd.crosstab(df['纹饰'], df['表面风化'])  # 生成纹饰与表面风化的交叉表
```

```python
# 绘制玻璃类型与风化的条形图
type_wind_counts = df.groupby(['类型', '表面风化']).size().unstack()
type_wind_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('玻璃类型与风化的条形图')
plt.xlabel('玻璃类型')
plt.ylabel('计数')
plt.legend(title='表面风化')
plt.show()
```

```python
# 绘制颜色与风化的条形图
color_wind_counts = df.groupby(['颜色', '表面风化']).size().unstack()
color_wind_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('颜色与风化的条形图')
plt.xlabel('颜色')
plt.ylabel('计数')
plt.legend(title='表面风化')
plt.show()
```

```python
# 绘制纹饰与风化的条形图
pattern_wind_counts = df.groupby(['纹饰', '表面风化']).size().unstack()
pattern_wind_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('纹饰与风化的条形图')
plt.xlabel('纹饰')
plt.ylabel('计数')
plt.legend(title='表面风化')
plt.show()
```
